```python
from bs4 import BeautifulSoup
import requests
from warnings import filterwarnings 
filterwarnings("ignore")
```

```python
url="https://timesofindia.indiatimes.com/news"
```

```python
r=requests.get(url)
data=r.content
soup=BeautifulSoup(data,"html.parser")
```

```python
len(url)
```

```python
urllist=[]
for i in range(0,40,1):
    x=url+str(i)
    urllist.append(x)
```

```python
urllist
```

```python
headlines=[]
for j in urllist:
    soup=BeautifulSoup(data,"html.parser")
    for i in soup.find_all("span",class_="w_tle"):
        headlines.append(i.find("a")["title"])
```

```python
headlines
```

```python
head=[]
import re
for i in headlines:
    rev = i.lower()
    head.append(re.sub("[^a-zA-Z0-9 ]","",rev))
```

```python
len(head)
```

```python
def tokenize(A):
    from nltk.tokenize import word_tokenize
    A1 = word_tokenize(A)
    return A1
```

```python
import nltk
nltk.download('punkt')
```

```python
head1= tokenize(str(head))
```

```python
head1
```

```python
head=[]
import re
for i in head1:
    rev = i.lower()
    head.append(re.sub("[^a-zA-Z ]","",rev))
```

```python
head
```

```python
def stopword(A3):
    from nltk.corpus import stopwords
    sw = stopwords.words("english")
    A4 = []
    for i in A3:
        if i not in sw:
            A4.append(i)
    return A4
```

```python
swhead = stopword(head)
len(swhead)
```

```python
from nltk.stem import LancasterStemmer
from nltk.stem import WordNetLemmatizer
ls=LancasterStemmer()
wnl=WordNetLemmatizer()

Q2=[]
for j in range(0,len(swhead)):
    lem=""
    for i in swhead[j].split(" "):
        lem = lem + wnl.lemmatize(ls.stem(i))
    Q2.append(lem) 
```

```python
#Q2[0:50]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(Q2).toarray()
```

```python
q1 = tfidf.get_feature_names()
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
model = km.fit(X)
```

```python
import pandas as pd
```

```python
B = pd.DataFrame(model.labels_,columns=["cluster"])
B.cluster.value_counts()
```

```python
model.cluster_centers_.argsort()[:,:-40:-1]
```

# we have to find Frequent word from the dataset

```python
common_words = model.cluster_centers_.argsort()[:,:-40:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' :         ' + ','.join(q1[word] for word in centroid))
```
