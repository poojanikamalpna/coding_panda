```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.preprocessing import MinMaxScaler
```

```python
A=pd.read_csv("C:/Users/Lenovo/Downloads/Electric_Production.csv")
```

```python
A.head()
```

```python
A.index=pd.to_datetime(A.DATE)
```

```python
A
```

```python
A=A.drop(labels=["DATE"],axis=1)
```

```python
A
```

```python
A.plot()
plt.xlabel("years")
plt.ylabel("electricity produce")
```

```python
sd=seasonal_decompose(A,model="multiplicative")
sd.plot()
```

```python
plt.plot(A)
plt.xlabel("years")
plt.ylabel("Electricity Production")
plt.plot(A.rolling(window=12).mean())
```

```python
mms=MinMaxScaler()
model=pd.DataFrame(mms.fit_transform(A),index=A.index,columns=A.columns)
```

```python
plt.plot(model)
plt.xlabel("years")
plt.ylabel("Electricity Production")
plt.plot(model.rolling(window=12).mean())
```

```python
B=np.log(A)
plt.plot(B)
plt.xlabel("years")
plt.ylabel("Electricity Production")
plt.plot(B.rolling(window=10).mean())
```

```python
C=B-B.rolling(window=12).mean()
plt.plot(C)
plt.xlabel("years")
plt.ylabel("Electricity Production")
plt.plot(C.rolling(window=10).mean())
```
