```python
def replacer(df):
    import pandas as pd
    Q=pd.DataFrame(df.isna().sum(),columns=["mv"])
    R=Q[Q.mv>0]
    for i in R.index:
        if (df[i].dtypes=="object"):
            mode=df[i].mode()[0]
            df[i]=df[i].fillna(mode)
        else:
            mean=df[i].mean()
            df[i]=df[i].fillna(mean)
        
```
