# Get All Article:

```python
articles=[]
def article_fetch(url):
    data1=requests.get(url)
    soup1=BeautifulSoup(data1.content,"html.parser")
    art=""
    for i in soup1.find_all("div",class_="td-module-meta-info"):
        art=art + str(i.text)
    return(art)
```

```python
articles
```

```python
for i in master_url_list:
    article=article_fetch(i)
    articles.append(article)
    
```

```python
art = []
for i in range(0,len(articles)):
    art.append(articles[i][56:len(articles[i])])
```

```python
dates=[]
for i in articles:
    dates.append(i[0:24])
```

```python
import pandas as pd
Q=pd.DataFrame([dates,headlines,art]).T
Q.columns=["Date","Headline","Articles"]
Q
```

# Preprocessing Article

```python
import re
ppd=[]
for i in art:
    q=i.lower()
    q=re.sub("[^a-zA-Z ]","",q)
    ppd.append(q)
```

```python
from nltk.stem import LancasterStemmer
ps=LancasterStemmer()

from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
E=[]
ppdf=[]
for j in range(0,len(ppd)):
    wnl=[]
    for i in ppd[j].split(" "):
        wnl.append(lem.lemmatize(ps.stem(i)))
        E.append(lem.lemmatize(ps.stem(i)))
        
    ppdf.append(wnl)    
```

```python
len(ppdf)
```

```python
Q.Articles=ppdf
```

```python
Q
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv=TfidfVectorizer()
X=tfv.fit_transform(E).toarray()
```

```python
tfv.vocabulary_
```

```python
X
```

```python
words=tfv.get_feature_names_out()
```

```python
words
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
model=km.fit(X)
```

```python
model.labels_
```

```python
Q1=pd.DataFrame([words,model.labels_]).T
```

```python
cw=model.cluster_centers_.argsort()
```
